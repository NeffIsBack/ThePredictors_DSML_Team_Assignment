```python
import pandas as pd
```

```python
df_Q1 = pd.read_csv('Divvy_Trips_2017_Q1.csv')
df_Q2 = pd.read_csv('Divvy_Trips_2017_Q2.csv')
df_Q3 = pd.read_csv('Divvy_Trips_2017_Q3.csv')
df_Q4 = pd.read_csv('Divvy_Trips_2017_Q4.csv')
```

```python
df_Q1['start_time'] = pd.to_datetime(df_Q1.start_time)
df_Q2['start_time'] = pd.to_datetime(df_Q2.start_time)
df_Q3['start_time'] = pd.to_datetime(df_Q3.start_time)
df_Q4['start_time'] = pd.to_datetime(df_Q4.start_time)
```

```python
df_Q1['end_time'] =  pd.to_datetime(df_Q1.end_time)
df_Q2['end_time'] =  pd.to_datetime(df_Q2.end_time)
df_Q3['end_time'] =  pd.to_datetime(df_Q3.end_time)
df_Q4['end_time'] =  pd.to_datetime(df_Q4.end_time)
```

```python
df_Q1.sort_values(by= 'start_time', inplace = True)
df_Q2.sort_values(by= 'start_time', inplace = True)
df_Q3.sort_values(by= 'start_time', inplace = True)
df_Q4.sort_values(by= 'start_time', inplace = True)
```

```python
df_all= pd.concat([df_Q1, df_Q2, df_Q3, df_Q4], ignore_index = True)

```
