```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import seaborn as sns

import folium
from folium import plugins
from folium.plugins import HeatMap
from datetime import datetime #for working with times objects
from datetime import timedelta #for working with times objects
import math
import random
```

```python
chg_2017 = pd.read_csv("chicago_2017.csv")
chg_2017.sort_values(["bike_id","start_time"], inplace=True)
chg_2017.set_index(pd.DatetimeIndex(chg_2017["start_time"]),inplace=True)
chg_2017.drop("start_time", axis=1, inplace=True)

#chg_2017.info()
# Check if the dataframe is complete 
#chg_2017.count()
```

Fleet Usage during a day, a week and the year:

```python
def get_weekday (ts):
    return ts.weekday()

def get_month (ts):
    return ts.month()
```

```python
chg_2017.reset_index(inplace=True)
chg_2017["Weekday"]= chg_2017["start_time"].apply(lambda ts: get_weekday(ts))
chg_2017["Month"]= chg_2017["start_time"].apply(lambda x: get_month(x))

```
